```python
# Renameing Column Names
new_cols = {
    'timestamp': 'Time',
    'cnt': 'Count',
    't1': 'Temp Real (°C)',
    't2': 'Temp Feels (°C)',
    'hum': 'Humidity %',
    'wind_speed': 'Wind Speed (kph)',
    'weather_code': 'Weather',
    'is_holiday': 'Holiday',
    'is_weekend': 'Weekend',
    'season': 'Season',
}

london_bikes.rename(new_cols, axis=1, inplace=True)

london_bikes.head()
```

```python
london_bikes['Humidity %'] = london_bikes['Humidity %'] / 100

london_bikes.head()
```

```python
# Mapping Season with actual Seasons
season = {
    '0.0': 'Spring',
    '1.0':'Summer',
    '2.0':'Autumn',
    '3.0':'Winter',
}

london_bikes['Season'] = london_bikes['Season'].astype('str').map(season)

london_bikes.head()
```

```python
# Mapping Weather with actual Weathers
weather = {
    '1.0': 'Clear',
    '2.0':'Scattered Clouds',
    '3.0': 'Broken Clouds',
    '4.0': 'Cloudy',
    '7.0':'Rain',
    '10.0':'Thunderstorm', 
    '26.0':'Snowfall',
}

london_bikes['Weather'] = london_bikes['Weather'].astype('str').map(weather)

london_bikes.head()
```
